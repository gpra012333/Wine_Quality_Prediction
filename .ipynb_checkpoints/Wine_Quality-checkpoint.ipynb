```python
X = wine_dataset.drop('quality',axis=1)
```

```python
Y = np.array(wine_dataset['quality']>=7).astype(int)
```

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(Train_X.shape)
print(Train_Y.shape)
print(Test_X.shape)
print(Test_Y.shape)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(Train_X, Train_Y)
```

```python
Train_Y_pred = model.predict(Train_X)
```

```python
Training_accuracy = accuracy_score(Train_Y_pred, Train_Y)
```

```python
print(Training_accuracy)
```

```python
Test_Y_pred = model.predict(Test_X)
```

```python
Testing_accuracy = accuracy_score(Test_Y,Test_Y_pred)
```

```python
print(Testing_accuracy)
```

```python
input_data = (7.9,0.32,0.51,1.8,0.341,17.0,56.0,0.9969,3.04,1.08,9.2,23)
input_data_nparray = np.asarray([input_data])
input_data_nparray = input_data_nparray.reshape(1,-1)
```
